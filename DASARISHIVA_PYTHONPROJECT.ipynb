```python
import pandas as pd
```

```python
# merging multiple worksheets in one dataset
df=pd.concat(pd.read_excel("C:\\Users\\Dasari Shiva\\Downloads\\data.xlsx",sheet_name=None),ignore_index=True)
df.head()
```

```python
df.shape
```

```python
#finding null values
df.isnull().sum()
```

```python
#removing column which has null values
df2 = df.drop("alt_sku", axis =1,)
df2
```

# IN STOCK

```python
df3=df2[df2['status']=='In Stock']
df3
```

```python
# creating total revenue column
df3["Total_revenue"]=df3["qty"]*df3["price"]
```

```python
df3.columns
```

```python
df3.head()
```

```python
# 1. Which are the top Manufacturers with price, quantity?

Top_man=pd.DataFrame(df3.groupby(['manufacturer']).sum().sort_values('Total_revenue', ascending=False))
(Top_man.drop(['category','qty','price'], axis = 1))
```

```python
# 2. Which are the most profitable products?

Profit_products=pd.DataFrame(df3.groupby(['part_number']).sum().sort_values('Total_revenue', ascending=False))
(Profit_products.drop(['category','qty','price'], axis = 1))
```

```python
# 3. Which is the product where you have seen a decrease in the price?
prices=pd.DataFrame(df3.groupby(['part_number']).sum().sort_values('price', ascending=True))
(prices.drop(['category','qty','Total_revenue'], axis = 1))
```

```python
# decrease in price
price_trend1=pd.DataFrame(prices.drop(['category','qty','Total_revenue'], axis = 1))
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
price_trend1.plot()
```

```python
price_trend1.min()
```

```python
# 3. Which is the product where you have seen a increase in the price?
prices=pd.DataFrame(df3.groupby(['part_number']).sum().sort_values('price', ascending=False))
(prices.drop(['category','qty','Total_revenue'], axis = 1))
```

```python
price_trend=pd.DataFrame(prices.drop(['category','qty','Total_revenue'], axis = 1))
```

```python
price_trend.shape
```

```python
#increase in price
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
price_trend.plot()
```

```python
price_trend.max()
```

```python
# 5. Which is the top category by price and quantity?
Top_cat=pd.DataFrame(df3.groupby(['category']).sum().sort_values('Total_revenue', ascending=False))
(Top_cat.drop(['price','qty',], axis = 1))
```
